```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd 
import random
import statistics
import seaborn as sns 
import collections
import statsmodels.api as sm
import statsmodels.formula.api as smf
import scipy.stats as stats
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter, AutoMinorLocator)
```

```python
np.random.seed(0)

pop_size = 1000000
X = np.random.normal(100,1,size=pop_size)
E = np.random.normal(0,1,size=pop_size)
alpha = 0
beta = 0.5
Y = alpha + (beta*X) + E

regressor = sm.add_constant(X)
results = sm.OLS(Y, regressor).fit()

fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.grid()
ax.scatter(X,Y, marker=".")
#ax.set_aspect('equal')

ax.plot(X, (alpha + (beta*X)), color="red")
ax.plot(X, (results.params[0] + (results.params[1]*X)), color="orange", lw=3)



ax_2 = fig.add_axes([0,-1.1,2,1])
ax_2.grid()
ax_2.scatter(X,Y, marker=".")

ax_2.plot(X, (alpha + (beta*X)), color="red")
ax_2.plot(X, (results.params[0] + (results.params[1]*X)), color="orange", lw=3)
```
